```python
# 2 Create universe and then train/dev/test from universe
import datefinder
import random
import time
import string

def genuuid(prefix):
    return prefix + '_' + ''.join([random.choice(string.digits + string.letters) for _ in range(16)])

def getEvents(data_i):
    events = list()
    for j in range(len(data_i['turns'])):
        event = dict()
        event['data'] = data_i['turns'][j]['text']
        event['agent'] = 0 if data_i['turns'][j]['author'] =='user' else 1
        event["start_time"] = None
        event["time"] = time.time()
        event["action"] = "message"
        events.append(event)
        if(data_i['turns'][j]['author'] == 'wizard'):
            active_frame = data_i['turns'][j]['labels']['active_frame']
            #print data_i['turns'][j]['labels']['frames'][active_frame-1]['info'].keys()
            #print "----"
    return events

def getOutcome(data_i):
    is_failed = False 
    for j in range(len(data_i['turns'])):
        if(data_i['turns'][j]['author'] == 'user'):
            active_frame = data_i['turns'][j]['labels']['active_frame']
            if('NO_RESULT' in data_i['turns'][j]['labels']['frames'][active_frame-1]['info']):
                is_failed = data_i['turns'][j]['labels']['frames'][active_frame-1]['info']['NO_RESULT'][0]['val']
    outcome = dict()
    outcome['reward'] = int(not is_failed and data_i['labels']['wizardSurveyTaskSuccessful'])
    return outcome

def parse(info):
    to_delete = ['gym','vicinity','count_seat','n_children','n_adults','impl_anaphora','count_amenities','dst_city_ok','str_date_ok','end_date_ok',
                 'REQUEST_ALTS','count_dst_city','seat_ok','ref_anaphora','count_category','n_adults_ok','budget_ok',
                 'HEARMORE','intent_ok','NO_RESULT','action','intent','flex','amenities','count','count_name','downtown']
    for i in to_delete:
        if i in info:
            del info[i]
    if 'REJECTED' in info:
        return {}
    entry = dict()
    datemapper = {'str_date':'leaving_departure_date','end_date':'returning_departure_date','arr_time_dst':'leaving_arrival_time',
                  'dep_time_dst':'returning_departure_time','dep_time_or':'leaving_departure_time','arr_time_or':'returning_arrival_time'}
    for key in info:
        if not info[key][0]['val'] or info[key][0]['val']=='$MAX' or info[key][0]['val']=='$MIN' or info[key][0]['val']=='-1':
            continue
            
        #################
        # TO REMOVE LATER
        if(isinstance(info[key][0]['val'], unicode)):
            if ( info[key][0]['val'].startswith('$GT') or info[key][0]['val'].startswith('$LT')):
                
                continue
        #################
        
        if key in datemapper:
            matches = datefinder.find_dates(string.replace(string.replace(info[key][0]['val'],' am','am'),' pm','pm'))
            for match in matches:
                if datemapper[key].endswith('time'):
                    entry[datemapper[key]] = match.time()
                else:
                    entry[datemapper[key]] = match.date().replace(year=2016)
        else:
            if(isinstance(info[key][0]['val'], unicode)):
                if info[key][0]['val'].encode('utf-8').lower() in place_mapper:
                    entry[key] = place_mapper[info[key][0]['val'].encode('utf-8').lower()]
                else:
                    entry[key] = info[key][0]['val'].encode('utf-8').lower()
                
            else:
                entry[key] = info[key][0]['val']
    
    return entry

def getkbs(data_i):
    kbs = list()
    for j in range(len(data_i['turns'])):
        backj = len(data_i['turns'])-j-1
        if(data_i['turns'][backj]['author'] == 'user'):
            frames = data_i['turns'][backj]['labels']['frames']
            for k in range(len(frames)):
                kbs.append(parse(frames[k]['info']))
            break
    return kbs


def getattributes(kbs):
    union_attributes = list()
    for entry in kbs:
        if(entry):
            union_attributes = list(set().union(entry.keys(),union_attributes))
    attributes = list()
    for attr in union_attributes:
        obj = dict()
        obj['value_type'] = attr
        obj['name'] = attr
        obj['unique'] = False
        attributes.append(obj)
    return attributes

def getEntries(X_train):
    # iterate over each entry
    arr = list()
    for i in range(len(data)):
        entry = dict()
        
        entry['uuid'] = genuuid('S')
        events = getEvents(data[i])
        entry['events'] = events
        
        outcome = getOutcome(data[i])
        entry['outcome'] = outcome

        agents = getAgents()
        entry['agents'] = agents
        
        entry['scenario'] = dict()
        entry['scenario_uuid'] = genuuid('C')
        kbs = getkbs(data[i])
        entry['scenario']['kbs'] = list()
        entry['scenario']['kbs'].append(kbs)
            
        attributes = getattributes(kbs)
        entry['scenario']['attributes'] = attributes  
    
        entry['scenario']['uuid'] = genuuid('S')

        X_train.append(entry)
    return X_train

def getAgents():
    return({
        "1": "wizard",
        "0": "user"
    })
X_train = []
X_train = getEntries(X_train)

# Write to file
with open('universe.json', 'w') as file:
    datawrite = (json.dumps(X_train[1:100],indent=4,default=str)) # Create a subset for dev
#      datawrite = (json.dumps(X_train,indent=4,default=str))
    file.write(datawrite + '\n')
%run /Users/16521/Desktop/nlpproj/cocoa-flights/src/scripts/split_dataset.py --example-paths universe.json --output-path ./
```
